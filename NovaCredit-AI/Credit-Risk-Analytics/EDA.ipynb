```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("banking.db")

query = """
SELECT city,
       COUNT(*) AS total_customers
FROM customers
GROUP BY city
ORDER BY total_customers DESC;
"""

result = pd.read_sql(query, conn)

conn.close()

result
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("banking.db")

query = """
SELECT customer_id,
       age,
       city,
       income,
       credit_score
FROM customers
WHERE credit_score < 600
LIMIT 10;
"""

result = pd.read_sql(query, conn)

conn.close()

result
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("banking.db")

query = """
SELECT customer_id,
       age,
       city,
       credit_score,
       late_payments,
       income
FROM customers
WHERE credit_score < 600
AND late_payments > 5
LIMIT 10;
"""

result = pd.read_sql(query, conn)

conn.close()

result
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("banking.db")

query = """
SELECT city,
       COUNT(*) AS total_customers,
       SUM([default]) AS total_defaults,
       ROUND(AVG([default])*100, 2) AS default_rate
FROM customers
GROUP BY city
ORDER BY default_rate DESC;
"""

result = pd.read_sql(query, conn)

conn.close()

result
```

```python
import matplotlib.pyplot as plt

# Average credit score grouped by default status
avg_credit = df.groupby("default")["credit_score"].mean()

plt.figure(figsize=(6,4))
plt.bar(avg_credit.index.astype(str), avg_credit.values)

plt.xlabel("Default (0 = No, 1 = Yes)")
plt.ylabel("Average Credit Score")
plt.title("Average Credit Score by Loan Default")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Select only numerical columns
numeric_df = df.select_dtypes(include=["int64", "float64"])

# Correlation matrix
corr = numeric_df.corr()

plt.figure(figsize=(10,8))

plt.imshow(corr)

plt.colorbar()

plt.xticks(
    range(len(corr.columns)),
    corr.columns,
    rotation=90
)

plt.yticks(
    range(len(corr.columns)),
    corr.columns
)

plt.title("Correlation Heatmap")

plt.show()
```
